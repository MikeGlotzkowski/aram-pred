```python
secret = 'super#secure#p4sswrd'
import pandas as pd
import pymongo as pm
import numpy as np
client = pm.MongoClient(f'mongodb://super#mongo#admin:{secret}@127.0.0.1:27017/')
collection = client['aram_data']['raw_match_details']
df = pd.DataFrame(list(collection.find({'queueId': 450}))) # 450 is Aram queue; could also filter by 'gameMode': 'ARAM'

teams = df['teams'].apply(pd.Series)
participants = df['participants'].apply(pd.Series)

teams = teams.rename(columns = lambda x : 'team_' + str(x + 1))
participants = participants.rename(columns = lambda x : 'participant_' + str(x + 1))

df = pd.concat([df, teams, participants], axis=1)
```

```python
chunks = []

for y in range (1, 3):
    df_temp = pd.json_normalize(df[f"team_{y}"])
    df_temp.rename(columns=lambda x: f"team{y}_{x}", inplace=True)
    df = pd.concat([df, df_temp], axis = 1)
    
for y in range(1, 11):
    df_temp = pd.json_normalize(df[f"participant_{y}"])
    df_temp.rename(columns=lambda x: f"participant{y}_{x}", inplace=True)
    df = pd.concat([df, df_temp], axis = 1)
```
